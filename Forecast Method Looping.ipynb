```r
tsdf = ts(df,,start = c(2021,1),end = c(2021,39), frequency = 52)
head(tsdf)
```

```r
train <- window(tsdf, start = c(2021,1), end = c(2021,round(0.8*nrow(tsdf))))
test <- window(tsdf, start = c(2021,round(0.2*nrow(tsdf))))
```

```r
train_holt <- function(col) {
    rmse = accuracy(holt(train[,col]),test[,grep(col, colnames(train))])
    return (rmse[2,2])
}
```

```r
train_arima <- function(col) {
    rmse = accuracy(forecast(auto.arima(train[,col],stepwise = FALSE , approximation=FALSE)),test[,grep(col, colnames(train))])
    return (rmse[2,2])
}
```

```r
train_croston <- function(col) {
    rmse = accuracy(croston(ifelse(train[,col]<0,0,train[,col])),test[,grep(col, colnames(train))])
    return (rmse[2,2])
}
```

```r
train_ets <- function(col) {
    rmse = accuracy(forecast(ets(train[,col])),test[,grep(col, colnames(train))])
    return (rmse[2,2])
}
```

```r
train_holt("B005MHDPBW")
train_arima("B005MHDPBW")
train_croston("B005MHDPBW")
train_ets("B005MHDPBW")
```

```r
data.frame(forecast(auto.arima(train[,'B07ZRG5PM1'],stepwise = FALSE, approximation=FALSE),h=10))[,1]
```

```r
negative_croston <- function(col){
    l = ifelse(train[,col]<0,0,train[,col])
    return (l)
}
```

```r
dong <- ncol(train)
cot <- 12
d = matrix(NA, nrow = dong, ncol=cot)
row = 2
for (i in colnames(train)){
    x <- train_holt(i)
    y <- train_arima(i)
    z <- train_croston(i)
    t <- train_ets(i)
    if (min(x,y,z,t) == x){
        d[grep(i, colnames(train)),row] = "holt"
        d[grep(i, colnames(train)),3:12] = data.frame(forecast(holt(train[,i]),h=10))[,1]
    } else if (min(x,y,z,t) == y){
        d[grep(i, colnames(train)),row] = "arima"
        d[grep(i, colnames(train)),3:12] = data.frame(forecast(auto.arima(train[,i],stepwise = FALSE, approximation=FALSE),h=10))[,1]
    } else if (min(x,y,z,t) == z){
        d[grep(i, colnames(train)),row] = "croston"
        d[grep(i, colnames(train)),3:12] = data.frame(forecast(croston(negative_croston(i)),h=10))[,1]
    } else {
        d[grep(i, colnames(train)),row] = "ets"
        d[grep(i, colnames(train)),3:12] = data.frame(forecast(ets(train[,i]),h=10))[,1]
    }
}
```
