```r
tsdf = ts(df,,start = c(2021,1),end = c(2021,33), frequency = 52)
head(tsdf)
```

```r
train <- window(tsdf, start = c(2021,1), end = c(2021,26))
test <- window(tsdf, start = c(2021,27))
```

```r
train_holt <- function(col) {
    rmse = accuracy(holt(train[,col], h=7),test[,grep(col, colnames(train))])
    return (rmse[2,2])
}
```

```r
train_arima <- function(col) {
    rmse = accuracy(forecast(auto.arima(train[,col],stepwise = FALSE , approximation=FALSE),h=7),test[,grep(col, colnames(train))])
    return (rmse[2,2])
}
```

```r
train_croston <- function(col) {
    rmse = accuracy(croston(train[,col], h=7),test[,grep(col, colnames(train))])
    return (rmse[2,2])
}
```

```r
train_ets <- function(col) {
    rmse = accuracy(forecast(ets(train[,col]),h=7),test[,grep(col, colnames(train))])
    return (rmse[2,2])
}
```

```r
train_holt("9E00")
train_arima("9E00")
train_croston("9E00")
train_ets("9E00")
```

```r
d = data.frame(sku=colnames(train),model=rep(0,746),f1=rep(0,746),f2=rep(0,746),f3=rep(0,746),f4=rep(0,746),f5=rep(0,746),f6=rep(0,746),f7=rep(0,746))
row = 2
for (i in colnames(train)){
    x <- train_holt(i)
    y <- train_arima(i)
    z <- train_croston(i)
    t <- train_ets(i)
    if (min(x,y,z,t) == x){
        d[grep(i, colnames(train)),row] = "holt"
        d[grep(i, colnames(train)),3:9] = data.frame(forecast(train_holt(i),h=7))[,1]
    } else if (min(x,y,z,t) == y){
        d[grep(i, colnames(train)),row] = "arima"
        d[grep(i, colnames(train)),3:9] = data.frame(forecast(train_arima(i),h=7))[,1]
    } else if (min(x,y,z,t) == z){
        d[grep(i, colnames(train)),row] = "croston"
        d[grep(i, colnames(train)),3:9] = data.frame(forecast(train_croston(i),h=7))[,1]
    } else {
        d[grep(i, colnames(train)),row] = "ets"
        d[grep(i, colnames(train)),3:9] = data.frame(forecast(train_ets(i),h=7))[,1]
    }
}
```
